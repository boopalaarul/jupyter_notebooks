```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as ppl
import scipy as sp

def phi_real(o,mean,stddev):
	return np.exp(-(o-mean)**2/(2*stddev**2))/np.sqrt(2*np.pi*stddev**2)
def phi_false(o,mean,stddev):
    return np.exp(-(o-mean)**2/(2*stddev**2))/np.sqrt(2*np.pi*stddev**2)

# np.linspace defines x axis of line graph; 1000 means 1000 intervals (resolution of plot); 190-210 or 150-250
os = np.linspace(190,210,1000)
# creates placeholder array of zeros to hold y axs values
phirs = np.zeros(len(os))
# fill out y values; no need for brackets, just tabs?
for t in range(len(os)):
	phirs[t] = phi_real(os[t],200,2)

sutairu_ = {}
sutairu = {
#"axes.titlesize" : 24,
"axes.labelsize" : 20,
#lines.linewidth : 3,
#"lines.markersize" : 10,
"xtick.labelsize" : 20,
"ytick.labelsize" : 20,
"figure.figsize" : (9, 5),
'font.family' : 'serif',
#'font.weight' : 50,
#'font.serif' : ['Times New Roman'],
}

with ppl.style.context(sutairu):
	ppl.plot(os,phirs)

#ppl.legend(loc = "best")
ppl.ylabel("Probability density")

#pl.locator_params(axis='y', nbins=6)
ppl.locator_params(axis='x', nbins=15)

#creates vertical line from botton of axes (0) to top (1) at x position
ppl.axvline(194,0,1,color='blue',label="+- 3 SD")
ppl.axvline(196,0,1,color='green',label='+- 2 SD')
ppl.axvline(198,0,1,color='red',label='+- 1 SD')
ppl.axvline(202,0,1,color='red',label='+- 1 SD')
ppl.axvline(204,0,1,color='green',label='+- 2 SD')
ppl.axvline(206,0,1,color='blue',label='+- 3 SD')
ppl.xlabel ("Observations (ot)")

ppl.show()
```
